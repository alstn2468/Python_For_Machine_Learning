X0 + X1 + X2 + X1^2 + X1X2 + X2^2

### Example
- f 함수
\begin{align}
x * sin(x^2) + 1 반환
\end{align}

- sample 함수<br/>
    + 임의의 값을 넣어 랜덤 값 생성
    
- linspace 함수<br/>
    + a 부터 b까지 사이의 size가 n개인 수 균등하게 생성

```python
def f(size):
    x = np.linspace(0, 5, size)
    y = x * np.sin(x ** 2) + 1
    return (x, y)

def sample(size):
    x = np.linspace(0, 5, size)
    y = x * np.sin(x ** 2) \
        + 1 + np.random.randn(x.size) * 0.5
    return (x, y)
```

```python
f_x, f_y = f(1000)
plt.plot(f_x, f_y)
X, y = sample(1000)
plt.scatter(X, y, s=3, c='black')
plt.show()
```

```python
X
```

```python
X.shape, y.shape
```

```python
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)
X.shape, y.shape
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X, y)
```

```python
f_x, f_y = f(1000)
plt.plot(f_x, f_y)
plt.scatter(X.flatten(), y.flatten(), s=3, c='black')
plt.plot(X.flatten(), lr.predict(X).flatten())
plt.show()
```

```python
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)
X_poly[:10]
```

```python
lr = LinearRegression(fit_intercept=False)
lr.fit(X_poly, y)
```

```python
f_x, f_y = f(1000)
plt.plot(f_x, f_y)
plt.scatter(X.flatten(), y.flatten(), s=3, c='black')
plt.plot(X.flatten(), lr.predict(X_poly).flatten())
plt.show()
```

```python
poly_features = PolynomialFeatures(degree=16)
X_poly = poly_features.fit_transform(X)
X_poly[:3]
```

```python
lr = LinearRegression(fit_intercept=False)
lr.fit(X_poly, y)
f_x, f_y = f(1000)
plt.plot(f_x, f_y)
plt.scatter(X.flatten(), y.flatten(), s=3, c='black')
plt.plot(X.flatten(), lr.predict(X_poly).flatten())
plt.show()
```

### 하고있는 작업
- parameter search
    + degree를 바꾸며 진행
    + 실험적으로 접근
    
#### rmse(Root Mean Square Error)
- 추정 값 또는 모델이 예측한 값과 실제 환경에서<br/>
관찰되는 값의 차이를 다룰 때 흔히 사용하는 측도

```python
def rmse(predctions, targets):
    return np.sqrt(((predctions - targets) ** 2).mean())
```

- Lasso : L1<br/>
- Ridge : L2<br/>
- 가장 성능이 좋은 모델 : rmse의 값이 가장 작은 것

```python
poly_range = list(range(10, 50))
rmse_lr_list = []
rmse_lasso_list = []
rmse_ridge_list = []

from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge

for poly_value in poly_range:
    poly_features = PolynomialFeatures(degree=poly_value)
    X_poly = poly_features.fit_transform(X)
    
    lr = LinearRegression(fit_intercept=False)
    lr.fit(X_poly, y)
    
    rmse_lr_list.append(rmse(lr.predict(X_poly), y))

    lasso = Lasso(fit_intercept=False)
    lasso.fit(X_poly, y)
    rmse_lasso_list.append(rmse(lasso.predict(X_poly), y))
    
    ridge = Ridge(fit_intercept=False)
    ridge.fit(X_poly, y)
    rmse_ridge_list.append(rmse(ridge.predict(X_poly), y))
```

```python
from pandas import DataFrame

data = {
    "poly_range":poly_range,
    "lr_rmse":rmse_lr_list,
    "lasso_rmse":rmse_lasso_list,
    "ridge_rmse":rmse_ridge_list,
}

df = DataFrame(data).set_index("poly_range")
df
```

### 어떻게 최적화 하는가?
- **RMSE**의 최소값을 찾자
- **Ridge**, **Lasso**, **LR** 모두 써보자
- Degree를 10 ~ 50까지 써보자
- 결과를 **한눈에 정리**해보자

```python
plt.plot(poly_range, df["ridge_rmse"], label="ridge")
plt.plot(poly_range, df["lr_rmse"], label="lr")
plt.plot(poly_range, df["lasso_rmse"], label="lasso")
plt.legend()
plt.show()
```

### L1(Lasso)
- Unstable Solution<br/>
- Always on Solution<br/>
- Sparse Solution<br/>
- Feature Selection

### L2(Ridge)
- Stable Solution<br/>
- Only one Solution<br/>
- Non-sparse Solution

### L1의 값이 비슷한 이유
- L1(Lasso)의 경우 의미없는 데이터들을 0으로 바꾸기 때문

```python
df.min()
```

확인 결과 **Ridge**가 가장 작다.

```python
df["ridge_rmse"].sort_values().head()
```

가장 작은 Ridge의 **최솟값** 정렬하여 확인<br/>
`degree`가 `22`일때 가장 값이 적다.

```python
poly_features = PolynomialFeatures(degree=22)
X_poly = poly_features.fit_transform(X)
ridge = Ridge(fit_intercept=False)
ridge.fit(X_poly, y)
```

```python
f_x, f_y = f(1000)
plt.plot(f_x, f_y)
plt.scatter(X.flatten(), y.flatten(), s=3, c="black")
plt.plot(X.flatten(), ridge.predict(X_poly).flatten())
plt.show()
```
